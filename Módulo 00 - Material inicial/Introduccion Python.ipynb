### Diccionarios

Los diccionarios se utilizan para almacenar valores de datos en pares clave: valor.

Un diccionario es una colección ordenada (a partir de Python 3.7), modificable y que no admite duplicados.

**Operaciones:**
* Acceder a elementos: 
      thisdict = {"brand": "Ford",
                  "model": "Mustang",
                  "year": 1964}
      x = thisdict["model"]

      print(x)

    Output:
      
      Mustang

* El método keys () devolverá una lista de todas las claves en el diccionario

      x = thisdict.keys()

    Output:
      
      dict_keys(['brand', 'model', 'year'])

* El método values ​​() devolverá una lista de todos los valores en el diccionario

      x = thisdict.values()

    Output:
      
      dict_values(['Ford', 'Mustang', 1964])

* Para determinar si una clave específica está presente en un diccionario, use la palabra clave **in**
      if "model" in thisdict:
          print("Yes, 'model' is one of the keys in the thisdict dictionary")

    Output:
      Yes, 'model' is one of the keys in the thisdict dictionary

* Puede cambiar el valor de un elemento específico haciendo referencia a su nombre de clave:
      thisdict["year"] = 2018
      print(thisdict)

    Output:
      {"brand": "Ford", "model": "Mustang", "year": 2018}

* La adición de un elemento al diccionario se realiza utilizando una nueva clave de índice y asignándole un valor
      thisdict["color"] = "red"
      print(thisdict)

    Output:
      {"brand": "Ford", "model": "Mustang", "year": 2018, "color": "red"}

* El método pop() elimina el elemento con el nombre de clave especificado
      thisdict.pop("model")
      print(thisdict)

    Output:
      {'brand': 'Ford', 'year': 2018}


* El método popitem () elimina el último elemento insertado

      thisdict =  {"brand": "Ford",
                   "model": "Mustang",
                   "year": 1964}

                  thisdict.popitem()
                  print(thisdict)
    
    Output:
      {'brand': 'Ford', 'model': 'Mustang'}


* Al recorrer un diccionario, el valor de retorno son las claves del diccionario, pero también existen métodos para devolver los valores.

      for x in thisdict:
          print(x)

    Output:
      brand
      model
      year

* Imprimir todos los valores en el diccionario, uno por uno

      for x in thisdict:
          print(thistdict[x])

* Se puede recorrer las claves y los valores mediante el método items():

      for x, y in thisdict.items():
          print(x, y)

    Output:
      brand Ford
      model Mustang
      year 1964

**Subconjuntos de filas**

Una gran parte de la ciencia de datos consiste en encontrar qué partes de su conjunto de datos son interesantes. Una de las técnicas más simples para esto es encontrar un subconjunto de filas que coincida con algunos criterios. Esto a veces se conoce como filtrar filas o seleccionar filas.

Hay muchas formas de crear un subconjunto de un DataFrame, quizás la más común es usar operadores relacionales para devolver Verdadero o Falso para cada fila, luego pasarlo entre corchetes.

      df[df["Cantidad"] > 5]
      df[df["Cliente"] == "Juan"]
Puede filtrar por varias condiciones a la vez utilizando el operador **&**.

      df[(df["Cantidad"] > 5) y (df["Cliente"] == "Juan")]

**Agrupaciones de datos**

Ahora bien, si queremos agrupar los datos de alguna forma para mostrarlos o bien utilizarlos, necesitamos utilizar un metodo llamado .groupby() que nos va a permitir ademas de agruparlos realizar una operacion sobre los valores.

La sintaxis resumida para entender el groupby es asi, de todas formas recomendamos revisar la documentacion para mas detalle y posibilidades:

      df.groupby('columna agrupadora')['columna a operar matematicamente'].funcion()

Veamos ejemplos para interpretar el codigo:

```python
# supongamos que queremos agrupar la cantidad total de ventas por cada tipo de fruta

primer_df.groupby('Fruta')['Cantidad'].sum()
```

```python
# ahora supongamos que queremos hacer lo mismo pero por cliente
 
primer_df.groupby('Cliente').agg(cantidad = ("Cantidad", "sum"))
```

```python
# ahora tenemos ganas de ver la venta promedio de cada fruta

df2 = primer_df.groupby('Fruta')['Cantidad'].mean()
```

```python
# por ultimo tenemos ganas de saber la suma de frutas por persona comprada
 
df1 = primer_df.groupby(['Cliente', 'Fruta'])['Cantidad'].sum()  # si prestan atencion al groupby, podran ver que en vez de un string, pasamos una lista
```
